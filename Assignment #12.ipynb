```python
#importing numpy package
import numpy as np

#importing pandas package
import pandas as pd

#importing matplotlib package
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading delivery dataframe
delivery=pd.read_csv('deliveries.csv')

#reading matches dataframe
data=pd.read_csv('matches.csv')
```

```python
#merging two dataframes
new=delivery.merge(data,left_on='match_id',right_on='id')
```

```python
# Problem 1: Create function to plot 2d line plot depicting season by season performance of a batsman.

#creating season_batsmanperf() to plot above mentioned graph
def season_batsmanperf(batsman):
    
    #finding columns associated with the required batsman
    a=new[new['batsman']==batsman]
    
    #calculating runs per season of the required batsman
    b=a.groupby('season')['batsman_runs'].sum()
    
    #plotting required graph
    b.plot(kind='line',marker='D')
```

```python
# Problem 2: Create function to show pie chart suggesting percentage of runs scored with scoring shots of particular batsman.

#creating batsman_scoreshots() to show the pie chart mentioned above
def batsman_scoreshots(batsman):
    
    #finding columns associated with the required batsman
    a=new[new['batsman']==batsman]
    
    #calculating runs scored with particular scoring shots of the required batsman
    b=a.groupby('batsman_runs')['batsman_runs'].sum()
    
    #plotting required pie chart
    b.plot(kind='pie',autopct='%0.2f%%',shadow='True')
```

```python
#calling above function
batsman_scoreshots('CH Gayle')
```
