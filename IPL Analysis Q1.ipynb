```python
##1. Find out top 3 players(only top 3 no more) who have won most number of match of 
##the match award when the match was played in Mumbai.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('matches.csv')
```

```python
mask1=data.city=='Mumbai'
data[mask1].player_of_match.value_counts().head(3)
```

```python
def top_batsman(city1):
    mask2=data.city==city1
    return data[mask2].player_of_match.value_counts().head(3).plot(kind="pie",figsize=(8,8))
```

```python
top_batsman("Mumbai")
```

```python
top_batsman('Hyderabad')
```
